```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns


# Language categories
languages = ['French', 'Kannada']

# Accuracy values: one tuple per model (BLOOM 560M, BLOOM 1.7B, XLM-R Base)
model_names = ['BLOOM 560M', 'BLOOM 1.7B', 'XLM-R']
Accuracy = {
    'BLOOM 560M': [0.7227, 0.5908],
    'BLOOM 1.7B': [0.7551, 0.6154],
    'XLM-R': [0.7483, 0.6880],
}
F1 = {
    'BLOOM 560M': [0.724, 0.5898],
    'BLOOM 1.7B': [0.7559, 0.6146],
    'XLM-R': [0.7481, 0.6862],
}
```

```python
def barplots(languages, model_names, metric, metric_name):
  colors = ['#427AA1', '#064789', '#9BC1BC']
  x = np.arange(len(languages))
  width = 0.18

  fig, ax = plt.subplots(figsize=(6, 4))


  for i, (model_name, acc) in enumerate(metric.items()):
      offset = width * i
      rects = ax.bar(x + offset, acc, width, label=model_name, color=colors[i])
      ax.bar_label(rects, padding=3, fmt='%.2f')

  # Labels and formatting
  if metric_name == 'Accuracy':
    y_label = 'Accuracy'
    title = 'Model-wise Accuracy Comparison across Languages'
  else:
    y_label = 'F1 Score'
    title = 'Model-wise F1 Score Comparison across Languages'

  ax.set_ylabel(y_label)
  ax.set_title(title)
  ax.set_xticks(x + width , languages)
  ax.set_ylim(0, 1.0)
  ax.legend(title='Model', loc='upper right', fontsize=6, title_fontsize=9)

  plt.tight_layout()
  plt.show()
```

```python
barplots(languages, model_names, Accuracy, 'Accuracy')
```

```python
barplots(languages, model_names, F1, 'F1')
```
