```python
import pandas as pd

# Load Oxford 3000 CSV (assumes you uploaded it to Colab or have a URL)
df = pd.read_csv('/oxford_3000.csv')

# Drop duplicate words, keeping the first occurrence
df_unique = df.drop_duplicates(subset=['word'])

# Get list of unique English words
english_words = df_unique['word'].tolist()
print(f"Total unique words: {len(english_words)}")
```

```python
english_words
```

```python
from deep_translator import GoogleTranslator
from tqdm import tqdm
import time

translations = []

for word in tqdm(english_words):
    try:
        fr = GoogleTranslator(source='en', target='fr').translate(word)
        hi = GoogleTranslator(source='en', target='hi').translate(word)
        kn = GoogleTranslator(source='en', target='kn').translate(word)
        translations.append({"en": word, "fr": fr, "hi": hi, "kn": kn})
        time.sleep(0.2)  # Avoid rate limiting
    except Exception as e:
        print(f"Error on '{word}': {e}")
        translations.append({"en": word, "fr": "", "hi": "", "kn": ""})
```

```python
translations[1]
```

```python
df_trans = pd.DataFrame(translations)
df_trans_cleaned = df_trans[(df_trans["fr"].str.strip() != "") & (df_trans["hi"].str.strip() != "") & (df_trans["kn"].str.strip() != "")]
df_trans_cleaned = df_trans_cleaned.reset_index(drop=True)
df_trans_cleaned.to_csv("oxford3000_translated.csv", index=False)
print("✅ Translations saved to 'oxford3000_translated.csv'")
df_trans_cleaned.head()
```
